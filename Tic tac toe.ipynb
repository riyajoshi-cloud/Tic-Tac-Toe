```python
def print_board(board):
    """Prints the 3x3 grid."""
    for row in board:
        print("|".join(row))
        print("-" * 5)

def check_winner(board):
    """Checks rows, columns, and diagonals for a win."""
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ": return board[i][0] # Rows
        if board[0][i] == board[1][i] == board[2][i] != " ": return board[0][i] # Cols
    if board[0][0] == board[1][1] == board[2][2] != " ": return board[0][0] # Diag 1
    if board[0][2] == board[1][1] == board[2][0] != " ": return board[0][2] # Diag 2
    return None

def play_game():
    board = [[" " for _ in range(3)] for _ in range(3)]
    player = "X"
    for turn in range(9):
        print_board(board)
        # Get input, check for valid integers and range
        try:
            move = int(input(f"Player {player}, enter 1-9 (top-left to bottom-right): ")) - 1
            row, col = divmod(move, 3)
            if board[row][col] != " ": raise ValueError
            board[row][col] = player
        except (ValueError, IndexError):
            print("Invalid move, try again.")
            continue
        
        if check_winner(board):
            print_board(board)
            print(f"Player {player} wins!")
            return
        player = "O" if player == "X" else "X"
    print_board(board)
    print("It's a tie!")

if __name__ == "__main__":
    play_game()
```
